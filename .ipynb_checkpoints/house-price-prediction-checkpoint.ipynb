# House Price Prediction

```python
import os

import numpy as np
import pandas as pd
import tensorflow as tf

import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("ggplot")

%matplotlib inline
%config InlineBackend.figure_formats = ("png", "retina")
colors = ['#e66101','#fdb863','#b2abd2','#5e3c99']
```

```python
os.listdir("./data")
```

```python
train_path = "./data/train.csv"
test_path = "./data/test.csv"

train = pd.read_csv(train_path)
test = pd.read_csv(test_path)
```

```python
print(train.columns)
print("-"*60)
print(train.shape)
```

```python
train.describe()
```

```python
with plt.style.context("ggplot"):
    plt.figure(figsize=(12, 5))
    plt.subplot(121)
    sns.distplot(train[['SalePrice']], kde=False)
    plt.xlabel("Sale Price")
    plt.axis([0, 8e5, 0, 180])
    plt.xticks(rotation=15)
    plt.title("Tail-heavy distribution")
    plt.subplot(122)
    sns.distplot(np.log(train[['SalePrice']]), kde=False)
    plt.xlabel("Log (sale price)")
    plt.axis([10, 14, 0, 180])
    plt.title("More normal after log transformation")
```

## Have a look at the variable correlations

```python
corr = train.select_dtypes(include=['float64', 'int64']).iloc[:, 1:].corr()

fig, ax = plt.subplots(figsize=(12, 12/1.4))
sns.set(font_scale=2)
sns.heatmap(corr, vmax=1, square=True, cmap="viridis")
plt.show()
```

```python
corr_list = corr['SalePrice'].sort_values(axis=0, ascending=False).iloc[1:]
corr_list
```

```python
corr_df = pd.DataFrame({"correlation": corr_list.values,
                              "features": corr_list.index}).sort_values(by="correlation")

with plt.style.context("ggplot"):
    fig, ax = plt.subplots(figsize=(8, 8*1.4))
    corr_df.plot.barh(x="features", y="correlation", 
                      ax=ax, color=colors[3], fontsize=12)
    ax.set_ylabel("")
```

We can also make use of the important measure of random forest or gradient boosting machine

```python
train_data = train.select_dtypes(include=["float64", "int64"]).dropna()
train_label = train_data.pop('SalePrice')
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_reg = RandomForestRegressor(n_estimators=50,
                               criterion='mae',
                               max_features=27,
                               random_state=42)
rf_reg.fit(train_data, train_label)
```

```python
y_pos = np.arange(corr_df.shape[0])
vars = list(corr_df["features"].values)

fig, axes = plt.subplots(1, 2, figsize=(10, 5*1.4))
fig.subplots_adjust(wspace=0.5)
axes[0].barh(y_pos, corr_df["correlation"],
             align="center", color=colors[0])
axes[0].set_yticks(y_pos)
axes[0].set_yticklabels(vars, fontsize=8)

axes[1].barh(np.arange(imp_data.shape[0])-1, imp_data["importance"],
             align="center", color=colors[3])
axes[1].set_yticks(y_pos)
axes[1].set_yticklabels(vars, fontsize=8)

plt.show()
```

```python
feature_importance = rf_reg.feature_importances_
imp_data = pd.DataFrame({"importance":feature_importance,
                         "features": train_data.columns}).sort_values(by="importance")

plt.style.use("ggplot")
fig, axes = plt.subplots(1, 2, figsize=(14, 6*1.2))
fig.subplots_adjust(wspace=0.3)
imp_data.plot.barh(y="importance", x="features",
                   ax=axes[0], color=colors[0],
                   fontsize=8)
corr_df.plot.barh(x="features", y="correlation",
                  ax=axes[1], color=colors[3],
                  fontsize=8)
```

```python
important_vars = imp_data["features"][:5]
df = pd.concat([train_label, train_data[important_vars]], axis=1)
sns.pairplot(df, size=2.5)
```
